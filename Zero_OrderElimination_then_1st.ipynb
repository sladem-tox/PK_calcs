```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    m = -4.25   # linear slope
    b = 20   # linear y-intercept
    C0 = 10  # initial value of exponential function
    k = 0.5  # decay constant of exponential function

    if x <= 5:
        return m*x + b
    else:
        return C0*np.exp(-k*(x-5)) + m*5 + b - C0  # adjusted exponential function

# Generate data for x and y values
x_values = np.linspace(0, 10, 100)
y_values = [f(x) for x in x_values]

# Plot the function
plt.plot(x_values, y_values)

# Add axis labels and a title
plt.xlabel(' Time (hours)')
plt.ylabel('Plasma Concentration (mg/L)')
plt.title('Function with Linear decline and Exponential Terminal Elimination')

# Show the plot
plt.show()
```
